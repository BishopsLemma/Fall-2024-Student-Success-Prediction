```python
# Import the usual suspects
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the data and print it
df = pd.read_csv("../data/newdata.csv")
df
```

```python
# Get the % of students that recieved a degree
# Iowa State's current graduation rate is 74.5
total_grads = df['DEG_CD'].sum()
print(total_grads)
grad_rate = total_grads/len(df['DEG_CD'])
print(grad_rate)
```

```python
# Get the list of course names for iteration purposes
course_names = df.columns.tolist()[5:]
course_names
```

```python
# We are going to create a new dataframe that will have the courses as the rows
# and eight columns - first four will be total number of 0,1,2,3's without graduating
# next four will be total number of 0,1,2,3's with graduating

# Make the new datafram
new_df = pd.DataFrame(columns=['Course','0_N','1_N','2_N','3_N','0_G','1_G','2_G','3_G'])

# For loop based on coursenames to make the rows
for n in course_names:
    list = [n]
    for j in [0,1]:
        for k in [0,1,2,3]:
            if (df[['DEG_CD',n]].values == [j,k]).all(axis=1).any() == True:
                list.append(df[['DEG_CD',n]].value_counts()[j,k]) 
                # this will tally the number of times j and k appear in DEG_CD and the course respectively
                # but will throw an error if there is nothing to tally, hence the if/else
            else:
                list.append(0)
    new_df.loc[len(new_df)] = list
```

```python
new_df
```

```python
# Sanity check - sum of each row should be 9779
new_df.sum(axis=1,numeric_only=True)
```

```python
# Add columns for the odds of graduating

new_df['0_O'] = new_df['0_G']/(new_df['0_N'] + new_df['0_G'])
new_df['1_O'] = new_df['1_G']/(new_df['1_N'] + new_df['1_G'])
new_df['2_O'] = new_df['2_G']/(new_df['2_N'] + new_df['2_G'])
new_df['3_O'] = new_df['3_G']/(new_df['3_N'] + new_df['3_G'])
```

```python
new_df
```

```python
# Let's get the mean of each column
print(new_df['0_O'].mean())
print(new_df['1_O'].mean())
print(new_df['2_O'].mean())
print(new_df['3_O'].mean())
```
